Assignment 3: Create a Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

# load a ipl dataset


```python
df = pd.read_csv("ipl_matches.csv.csv")
```

```python
# Set visual style
sns.set(style="whitegrid")
```

# Plot 1: Distribution of match winners

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, y='winner', order=df['winner'].value_counts().index, palette='Set3')
plt.title('Distribution of Match Winners')
plt.xlabel('Number of Matches Won')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

# Plot 2: Top 10 most winning teams

```python

plt.figure(figsize=(8, 6))
top_winners = df['winner'].value_counts().head(10)
sns.barplot(x=top_winners.values, y=top_winners.index, palette='coolwarm')
plt.title('Top 10 Most Winning Teams in IPL')
plt.xlabel('Wins')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

# Plot 3: Number of matches per season

```python

df['season'] = df['season'].astype(str)  # Ensure it's a string for ordering
season_counts = df['season'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.lineplot(x=season_counts.index, y=season_counts.values, marker='o')
plt.title('Number of Matches Played Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.tight_layout()
plt.show()
```

# Plot 4: Most common venues

```python

plt.figure(figsize=(10, 6))
venue_counts = df['venue'].value_counts().head(10)
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette='mako')
plt.title('Top 10 Most Played Venues')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.tight_layout()
plt.show()
```

# Plot 5: Most "Player of the Match" awards


```python

plt.figure(figsize=(10, 6))
pom_counts = df['player_of_match'].value_counts().head(10)
sns.barplot(x=pom_counts.values, y=pom_counts.index, palette='viridis')
plt.title('Top 10 Players with Most "Player of the Match" Awards')
plt.xlabel('Awards')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```

```python
# Pie chart of toss decisions (bat vs field)
toss_decision_counts = df['toss_decision'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(toss_decision_counts, labels=toss_decision_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Toss Decision Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Number of matches won per team per season
plt.figure(figsize=(14, 7))
season_team_wins = df.groupby(['season', 'winner']).size().reset_index(name='wins')
sns.lineplot(data=season_team_wins, x='season', y='wins', hue='winner', marker='o')
plt.title('Team Performance Trend Over Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
